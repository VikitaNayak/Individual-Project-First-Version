```python
import pandas as pd
from uszipcode import SearchEngine
search = SearchEngine(simple_zipcode=False)
from uszipcode import Zipcode
```

```python
file = open("zipcodes_results_3.csv", "a")
```

```python
with open("Longitude_latitude_crashes_unique.csv") as f:
    for line in f:
        read = line.split(',')
        hla = float(read[0])
        hlo = float(read[1])
        result = search.by_coordinates(hla,hlo,returns=1)
        for z in result:
            con_string = "%f,%f,%s" % (hla,hlo,z.zipcode)
            file.write(con_string + "\n")
            print(hla,hlo,z.zipcode)
            
```
